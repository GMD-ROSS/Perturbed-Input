### Autocorrelation 
#### First we compute the Autocovariance , then divide by the variance

```python
np.random.seed(0) 
products = []
mul = 0
time = np.arange(0,50)
signal = 10 + 0.02*time + np.random.normal(0,0.3,50)

mean = np.mean(signal)

for i in range(len(signal)):
    if i > 0:
       mul = (signal[i] - mean) * (signal[i-1] - mean)

       products.append(float(mul))

autocov = sum(products) / len(products)

variance = np.var(signal) 

autocorr = autocov / variance

print("The auto correlation is:", autocorr)
print("The auto covariance is:" , float(autocov))
print("The products are:", products)
```

#### Next we'll add more lag values and build an ACF graph of the data 

```python
np.random.seed(0) 
acf_values = []
mul = 0
max_lag = 10 
time = np.arange(0,50)

signal = 10 + 0.02*time + np.random.normal(0,0.3,50)

mean = np.mean(signal)
variance = np.var(signal) 

for lag in range(1, max_lag +1):
    products = [] 
    
    for i in range(len(signal)):
        if i >= lag:
           mul = (signal[i] - mean) * (signal[i-lag] - mean)

           products.append(float(mul))

    autocov = sum(products) / len(products)
    autocorr = autocov / variance

    acf_values.append(float(autocorr))

print(acf_values)

lags = list(range(1, max_lag+1))

plt.stem(lags ,acf_values) 
plt.xlabel("Lag")
plt.ylabel("ACF")
plt.title("ACF vs Lag")
plt.show 
```

```python
import pandas as pd 
import matplotlib.pyplot as pt

acf_values = []
differences = []
mul = 0
max_lag = 12

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"

df = pd.read_csv(url)

df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace = True)

signal = df['Passengers'].values

plt.figure()
plt.plot(df.index, signal)
plt.xlabel("Month")
plt.ylabel("Passengers (Thousands)")
plt.title("Passengers per Month")
plt.show()

mean = np.mean(signal)
variance = np.var(signal) 

for lag in range(1, max_lag +1):
    products = [] 
    
    for i in range(len(signal)):
        if i >= lag:
           mul = (signal[i] - mean) * (signal[i-lag] - mean) 
            
           products.append(float(mul))

    autocov = sum(products) / len(products)
    autocorr = autocov / variance

    acf_values.append(float(autocorr))

for i in range(len(signal)):
    if i > 0:
         difference = signal[i] - signal[i-1]
         differences.append(float(difference))


print("The ACF values are:", acf_values)
print("The differenced data values are:", differences)


lags = list(range(1, max_lag+1))

plt.figure()
plt.stem(lags,acf_values) 
plt.xlabsel("Lag")
plt.ylabel("ACF")
plt.title("ACF vs Lag")
plt.show() 

plt.figure()
plt.stem(df.index[1:],differences) 
plt.xlabel("Lag")
plt.ylabel("Differenced Data")
plt.title("Differenced Data Vs Month")
plt.show() 
```

##### We can see that even after first order differencing, the data is not fully stationary (variance is changing upward) although it looks to be centered at 0 which suggests a constant mean. 
##### Let's do a Dicky-fuller test to check stationarity mathematically:

```python
import pandas as pd 
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller

acf_values = []
differences = []
mul = 0
max_lag = 12

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"

df = pd.read_csv(url)

df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace = True)

signal = df['Passengers'].values

plt.figure()
plt.plot(df.index, signal)
plt.xlabel("Month")
plt.ylabel("Passengers (Thousands)")
plt.title("Passengers per Month")
plt.show()

mean = np.mean(signal)
variance = np.var(signal) 

for lag in range(1, max_lag +1):
    products = [] 
    
    for i in range(len(signal)):
        if i >= lag:
           mul = (signal[i] - mean) * (signal[i-lag] - mean) 
            
           products.append(float(mul))

    autocov = sum(products) / len(products)
    autocorr = autocov / variance

    acf_values.append(float(autocorr))

for i in range(len(signal)):
    if i > 0:
         difference = signal[i] - signal[i-1]
         differences.append(float(difference))


print("The ACF values are:", acf_values)
print("The differenced data values are:", differences)


lags = list(range(1, max_lag+1))

result = adfuller(signal)

print("ADF Statistic:" , result[0])
print("p-value:", result[1])
plt.figure()
plt.stem(lags,acf_values) 
plt.xlabel("Lag")
plt.ylabel("ACF")
plt.title("ACF vs Lag")
plt.show() 

plt.figure()
plt.stem(df.index[1:],differences) 
plt.xlabel("Lag")
plt.ylabel("Differenced Data")
plt.title("Differenced Data Vs Month")
plt.show() 
```

#### Since the p value is >> 0.05, we fail to reject ho and therefore the data is non-stationary

### Data set 2 - Monthly shampoo sales 
#### Let's use the Box-cox model identification workflow fully. 
#### As a reminder : 
1. Plot the data and identify unusual observations, look at patterns, trends, seasonality, outliers, is it stationary? 
2. Use a Box-Cox transformation to stabilize the variance if necessary. If not, move to step 6

##### The time plot shows an increasing trend, with a variance around the level which doesn't look too bad. Possible seasonality? But the ACF does not show a spike at lag 12, so we will assume no significant seasonality. The mean is clearly not constant. Analyzing the ACF, we can see that the trend moves downwards, indicating non-stationarity. Let's do a test to confirm this. We will use the Dicky-fuller test. 

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv"

df = pd.read_csv(url)
df.index = pd.date_range(start='2000-01-01', periods=len(df), freq='ME')

signal = df["Sales"].values

acf_values = []
mul = 0 

mean = np.mean(signal)
variance = np.var(signal) 

for lag in range(1, 13):
    products = [] 
    
    for i in range(len(signal)):
         if i >= lag:
           mul = (signal[i] - mean) * (signal[i-lag] - mean) 
           
           products.append(float(mul))

    autocov = sum(products) / len(products)
    autocorr = autocov / variance

    acf_values.append(float(autocorr))
    
lags = list(range(1,13))

result = adfuller(signal)

print("ADF Statistic:" , result[0])
print("p-value:", result[1])

print(mean)
print(variance)
```

##### The p value is 1, indicating the data is non stationary. Now lets stabilize the mean through differencing and check the mean. 

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv"

df = pd.read_csv(url)
df.index = pd.date_range(start='2000-01-01', periods=len(df), freq='ME')

signal = df["Sales"].values
differences = []
acf_values = []
mul = 0 

mean = np.mean(signal)
variance = np.var(signal) 

for lag in range(1, 13):
    products = [] 
    
    for i in range(len(signal)):
         if i >= lag:
           mul = (signal[i] - mean) * (signal[i-lag] - mean) 
           
           products.append(float(mul))

    autocov = sum(products) / len(products)
    autocorr = autocov / variance

    acf_values.append(float(autocorr))
    
for i in range(len(signal)):
    if i > 0:
         difference = signal[i] - signal[i-1]
         differences.append(float(difference))
    
lags = list(range(1,13))

differenced_mean = np.mean(differences) 

print(differenced_mean)

result = adfuller(differences)

print("ADF Statistic:" , result[0])
print("p-value:", result[1])

plt.figure()
plt.stem(df.index[1:],differences) 
plt.xlabel("Months")
plt.ylabel("Differenced Data")
plt.title("Differenced Data vs Months")
plt.show() 
```

#### The mean looks to be relatively constant over the sample period, and also doing the Fuller test results in a p value of 1.7998574141687034e-10, indicating that the data is stationary. Therefore, we can continue to the next step which is to plot the ACF and PACF of the differeneced data. From there we will determine possible candidate models. 

```python
from statsmodels.tsa.stattools import pacf

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv"

df = pd.read_csv(url)
df.index = pd.date_range(start='2000-01-01', periods=len(df), freq='ME')

signal = df["Sales"].values
differences = []
acf_values = []
pacf_values = []
mul = 0 

for i in range(len(signal)):
    if i > 0:
         difference = signal[i] - signal[i-1]
         differences.append(float(difference))

differenced_mean = np.mean(differences) 
differenced_variance = np.var(differences)
        
for lag in range(1, 13):
    products = [] 
    
    for i in range(len(differences)):
         if i >= lag:
           mul = (differences[i] - differenced_mean) * (differences[i-lag] - differenced_mean) 
           
           products.append(float(mul))

    autocov = sum(products) / len(products)
    autocorr = autocov / differenced_variance

    acf_values.append(float(autocorr))

pacf_values = pacf(differences, 12)
    
lags = list(range(1,13))

critical_value = 1.96 / np.sqrt(35) 

plt.figure()
plt.stem(lags,acf_values)
plt.axhline(critical_value,linestyle='--')
plt.axhline(-critical_value,linestyle='--')
plt.xlabel("Lag")
plt.ylabel("ACF")
plt.title("ACF vs Lag")
plt.show() 

plt.figure()
plt.stem(lags,pacf_values[1:])
plt.axhline(critical_value,linestyle='--')
plt.axhline(-critical_value,linestyle='--')
plt.xlabel("Lag")
plt.ylabel("PACF")
plt.title("PACF vs Lag")
plt.show() 
```

#### We see a cutoff for the ACF after lag 2, indicating the q value is assumed to be 2, and we see a cutoff after lag 1 in the PACF, indicating a p value of 1. And since our difference order is 1, we use 1 for d. So our estimate model is:
- ARIMA(1,1,2)

#### Our residuals look like white noise centered around 0, which is a good thing. Let's check the ACF of the residuals now:

```python
from statsmodels.graphics.tsaplots import plot_acf

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv"

df = pd.read_csv(url)
df.index = pd.date_range(start='2000-01-01', periods=len(df), freq='ME')

signal = df["Sales"].values

model = ARIMA(signal, order=(1,1,2))
fit = model.fit()

residuals = fit.resid

critical_value = 1.96 / np.sqrt(35)

plot_acf(residuals)
plt.axhline(critical_value,linestyle='--')
plt.axhline(-critical_value,linestyle='--')
plt.title("Residual ACF")
plt.show()
```

#### Our ACF for the residuals stays inside the critical values. So, the model looks good. Lets plot the forecast and compare it to our original signal. 

```python
from statsmodels.tsa.arima.model import ARIMA 

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv"

df = pd.read_csv(url)
df.index = pd.date_range(start='2000-01-01', periods=len(df), freq='ME')

signal = df["Sales"].values

model = ARIMA(signal, order=(1,1,2), trend = "t")

fit = model.fit()
fitted_values = fit.predict()

forecast_res = fit.get_forecast(steps=12)

forecast_values = forecast_res.predicted_mean

future_x = range(len(signal), len(signal) + 12)

plt.figure()
plt.plot(signal, label="Actual")
plt.plot(fitted_values, label="Fitted")
plt.plot(future_x, forecast_values, label="Forecast")
plt.legend()
plt.title("ARIMA Fit & Forecast vs Actual")
plt.show() 
```

```python
import numpy as np 
import matplotlib.pyplot as pt
import pandas as pd 
from statsmodels.tsa.stattools import acf

acf_values = []

critical_value = 1.96 / np.sqrt(3650)


df.index = pd.date_range(start='1981-01-01', periods=len(df), freq='D')

signal = df["Temp"]

mean = np.mean(signal)
variance = np.var(signal) 

for lag in range(1,366):
    products = [] 
    
    for i in range(len(signal)):
         if i >= lag:
           mul = (signal[i] - mean) * (signal[i-lag] - mean) 
           
           products.append(float(mul))

    autocov = sum(products) / len(products)
    autocorr = autocov / variance

    acf_values.append(float(autocorr))
    
lags = list(range(1,366))
    
plt.figure()
plt.stem(lags,acf_values)
plt.axhline(critical_value,linestyle='--')
plt.axhline(-critical_value,linestyle='--')
plt.title("ACF vs Lags")
plt.show()
```

#### Very high autocorrelation as expected (Temperature is cyclical/seasonal). Variance looks to be relatively stable. We need to implement seasonal differencing, and well check the ACF of the differenced data to confirm it's stationary

```python
import numpy as np 
import matplotlib.pyplot as pt
import pandas as pd 
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf

acf_values = []

critical_value = 1.96 / np.sqrt(3650)


df.index = pd.date_range(start='1981-01-01', periods=len(df), freq='D')

signal = df["Temp"]

signal_diff = signal.diff(365).dropna()

mean = np.mean(signal_diff)
variance = np.var(signal_diff) 


for lag in range(1,366):
    products = [] 
    
    for i in range(len(signal_diff)):
         if i >= lag:
           mul = (signal_diff[i] - mean) * (signal_diff[i-lag] - mean) 
           
           products.append(float(mul))

    autocov = sum(products) / len(products)
    autocorr = autocov / variance

    acf_values.append(float(autocorr))
    
lags = list(range(1,366))

pacf_values = pacf(signal_diff, 365)

plt.figure()
plt.stem(lags,pacf_values[1:])
plt.axhline(critical_value,linestyle='--')
plt.axhline(-critical_value,linestyle='--')
plt.title("Differenced PACF vs Lags")
plt.show()

plt.figure()
plt.plot(signal_diff.index,signal_diff)
plt.xlabel("Days")
plt.ylabel("Differenced Data")
plt.title("Differenced Data  vs Days")
plt.show()

plt.figure()
plt.stem(lags,acf_values)
plt.axhline(critical_value,linestyle='--')
plt.axhline(-critical_value,linestyle='--')
plt.title("Differenced ACF vs Lags")
plt.show()

print(mean)
print(variance)
```
