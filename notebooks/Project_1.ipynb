## Chapter 2 - The Forecaster's Toolbox

### Covariance 
- Measures the extent of the linear relationship between two variables
- Negative covariance means they move in opposite directions, positive covariance they move together (direction) 

### Correlation 
- Standardized covariance between -1 and 1
- Measures both strength of the relationship and direction

### Autocovariance 
- Gives the covariance of a process with itself at pairs of time points
- Not standardized

$$
c_k = \frac{1}{T} \sum_{t=k+1}^{T} (y_t - \bar{y})(y_{t-k} - \bar{y})
$$

- Where k is the lag by rk

### Autocorrelation 
- How similar the signal is to itself after it is shifted by k units
- Standardized
- For uncorrelated data, the autocorrelation should be close to 0

$$
r_k = \frac{c_k}{c_0}
$$

### Correlogram

- A plot of ACF values vs lag

- Seasonal ACF is not cyclical although it typically has a calendar based pattern
- Cyclical ACF does not follow a calendary based patter, but repeats itself 

### Residuals 

- Difference between observed value and its forecast based on all previous observations (error) 
- Uncorrelated, meaning there should be no pattern in the errors (random)
- If the errors are correlated, the model may have missed something
- Residuals should have a mean of 0 otherwise they are biased 
- Constant variance, normal distribution

### White Noise 

- Uncorrelated across time with zero mean and constant variance
- Residuals should look like white noise, a good diagnostic is to check the ACF of residuals through Portmanteau tests, Box-Pierce tests, Ljung-Box tests
- If all autocorrelation values are within the critical value limits, the data set is white noise. Critical values are found by: 

$$
\pm \frac{1.96}{\sqrt{T}}
$$

### Forecast Accuracy 

Mean Absolute Error (MAE)
$$
\mathrm{MAE} = \frac{1}{T} \sum_{t=1}^{T} \left| y_t - \hat{y}_t \right|
$$

Mean Squared Error (MSE)
$$
\mathrm{MSE} = \frac{1}{T} \sum_{t=1}^{T} \left( y_t - \hat{y}_t \right)^2
$$

Root Mean Squared Error (RMSE)
$$
\mathrm{RMSE} = \sqrt{ \frac{1}{T} \sum_{t=1}^{T} \left( y_t - \hat{y}_t \right)^2 }
$$

Mean Absolute Percentage Error (MAPE)
$$
\mathrm{MAPE} = \frac{100}{T} \sum_{t=1}^{T} \left| \frac{y_t - \hat{y}_t}{y_t} \right|
$$

Mean Absolute Scaled Error (MASE)

$$
\mathrm{MASE} = \frac{1}{T} \sum_{t=1}^{T} \frac{|y_t - \hat{y}_t|}{Q}
$$

Q is a scaling factor.

$$
Q = \frac{1}{T-1} \sum_{t=2}^{T} |y_t - y_{t-1}|
$$

Seasonal Data
$$
Q = \frac{1}{T-m} \sum_{t=m+1}^{T} |y_t - y_{t-m}|
$$

- We want the error to be low for each of the error computations

### Training and Test sets 

- Test set must NOT be used for any aspect of model development or calculation of forecast
- A model that is too well fit is not necessarily a good thing, we want to avoid capturing the noise, instead focusing on trends

## Chapter 3 - Exponential Smoothing

- Smoothing equation is:

$$
\ell_t = \alpha y_t + (1 - \alpha)\ell_{t-1}       
$$

- Forecase equation is the summation of all weighted past data: 
$$
\hat{y}_{t+h|t} = \sum_{j=1}^{t} \alpha (1-\alpha)^{t-j} y_j + (1-\alpha)^t \ell_0
$$
### State Space Form

#### Observation equation

- What we observe is what we expect + an error 

$$
y_t = \ell_{t-1} + e_t
$$

#### Computing the Error
$$
e_t = y_t - \hat{y}_t
$$

#### State Equation 

$$
\ell_t = \ell_{t-1} + \alpha e_t
$$

- Key idea is this: we predict $y_t$, observe a real value, subtract the prediction from the real value for an error and update the estimate
- Alpha and lo must be estimated, often by minimising the MSE

#### Methods for Modelling 

- Holt
- Exponential
- Damped trend
- Holt-Winters
  
##### ETS (Error, Trend, Seasonality) framework helps us decide which one is appropriate

- We have additive and multiplicative types of error
- Additive is constant where as multiplicative scales with the level
- Example, if we have ETS(A,A,M) this means we have an additive error, additive trend and multiplicative seasonality

### Chapter 6 - Making Time Series Stationary 

- If the size of the time series fluctuations change, ie. variations at different levels, transformations will be needed
- This essentially makes the variance constant and suitable for ARIMA and ETS 
- Transformation usually does not make a difference to forecasts, but has an effect on PI (Prediction Intervals) ie. uncertainty range
   
#### Box-Cox Transformations 

- For $\lambda = 0,$
  $w_t = \log(y_t)$
- For $\lambda \ne 0,$
  $w_t = \frac{y_t^\lambda - 1}{\lambda}$

- Choose a simple value of lambda
- Reverse transformations must be done to obtain forecasts on the original scale
- For $\lambda = 0$,
  $y_t = \exp(w_t)$
- For $\lambda \ne 0$,
  $y_t = (\lambda w_t + 1)^{1/\lambda}$

- Box-Cox transformation accompanied with an additive (must be additive) ETS model is better than ETS model alone

#### Stationarity 

- Stationary series are horizontal, does not depend on t
- Constant variance, constant mean, no patterns predictable (trend or seasonality) think noise around a fixed level
- How do we identify non-stationary series?
- The ACF of stationary data drops to zero relatively quickly
- The ACF of non-stationary data decreases slowly
- Typically for non-stationary data, the value of r1 is often large and positive

#### Ordinary Differencing 

- Differencing is used to stabilize the mean (ARIMA model needs stationary data) 
- The differenced series is the changes between each observation in the original series

#### Second-Order Differencing 

- We can difference the data a second time if it is not stationary after a single difference

#### Seasonal Differencing 

- Seasonal differenced series are closer to being stationary
- Can be combined with first differencing

- First differences are the changes between one observation and the next
- Seasonal differences are the changes between one year to the next 

#### Unit Root tests 

- Does the data wonder over time? If it does, it has a unit root, and therefore we need to difference it, if not, its stationary
- Dickey-Fuller test (null that data is non-stationary and non-seasonal)
- Kwiatkowksi-Phillips-Schmidt-Shin test (null that data is stationary and non-seasonal)

### Non-Seasonal ARIMA models 

#### Autoregressive models (AR)   

$y_t = c + \phi_1 y_{t-1} + \phi_2 y_{t-2} + \cdots + \phi_p y_{t-p} + e_t$

- $e_t$ is white noise

##### AR(1) model 

$y_t = c + \phi_1 y_{t-1} + e_t$

- Current value depends on last value and some random noise
- $\phi_1$ should be between -1 and 1
- If p > 1 , $\phi_2$ should be between -1 and 1, $\phi_1 + \phi_2 < 1$ , $\phi_2 - \phi_1 < 1$
  
#### Moving average models (MA) 

$y_t = c + e_t + \theta_1 e_{t-1} + \theta_2 e_{t-2} + \cdots + \theta_q e_{t-q}$

- The prediction is made from a linear combination of past errors 

#### ARIMA models 

##### Autoregressive Moving Average models (ARMA) 

$y_t = c + \phi_1 y_{t-1} + \cdots + \phi_p y_{t-p} + \theta_1 e_{t-1} + \cdots + \theta_q e_{t-q} + e_t$ 

- Predictors include lagged values of $y_t$ and lagged errors
- Conditions on coefficients ensure invertibility and stationarity 

##### Autoregressive Integrated Moving Average models (ARIMA)

- These combine ARMA model with differencing
  
- Backshift operators:

$
B y_t = y_{t-1}
$

$
B^2 y_t = y_{t-2}
$

- Substituting:
  
$
y_t = c + \phi_1 B y_t + \cdots + \phi_p B^p y_t
      + \theta_1 B e_t + \cdots + \theta_q B^q e_t + e_t
$

- Factor yt:

$
y_t - (\phi_1 B + \cdots + \phi_p B^p)y_t = c + (1 + \theta_1 B + \cdots + \theta_q B^q)e_t
$

- Simplifying: 

$
(1 - \phi_1 B - \cdots - \phi_p B^p)y_t = c + (1 + \theta_1 B + \cdots + \theta_q B^q)e_t
$

- Polynomial for phi

$
\phi(B) = 1 - \phi_1 B - \cdots - \phi_p B^p
$

- Polynomial for theta 

$
\theta(B) = 1 + \theta_1 B + \cdots + \theta_q B^q
$

- ARMA form: 

$
\phi(B) y_t = c + \theta(B) e_t
$

- By applying differencing, we transform to the ARIMA model:

$
\phi(B)(1 - B)^d y_t = c + \theta(B) e_t
$  

##### ARIMA(p,d,q) model 

- AR: p = order of autoregressive part
-  I: d = degree of first differencing involved
- MA: q = order of the moving average part
- White noise model: ARIMA(0,0,0)
- Random walk: ARIMA(0,1,0) with no constant
- Random walk with drift: ARIMA(0,1,0) with constant

- The higher the value of d, the more rapidly Forecast intervals increase in size
- For cyclic forecasts, p > 2 and restrictions on the coefficients are asserted
- For p = 2: 
$\phi_1^2 + 4\phi_2 < 0$ 

##### Partial autocorrelations 

- Parial autocorrelations measure the relationship between one data point and another when the affects of other time lags are removed
- ACF: How related are the datapoints overall
- PACF: Is there a direct connection between a set of datapoints

#### Estimation and order Selection

##### Maximum likelihood estimation (MLE) 

- Once we identify the model order, we need to estimate parameters c, $\phi_p$, $\theta_p$

#### Information Criteria

##### Akaike's Information Criterion (AIC): 
$
\mathrm{AIC} = -2 \log(L) + 2(p + q + k + 1)
$

- L is the likelihood of the data
- k=1 if c$\neq$0 and k = 0 if c = 0

##### Corrected AIC: 

$
\mathrm{AIC}_c = \mathrm{AIC} + \frac{2(p + q + k + 1)(p + q + k + 2)}{T - p - q - k - 2}
$

##### Bayesian Information Criterion: 

$
\mathrm{BIC} = \mathrm{AIC} + (p + q + k + 1)\log(T)
$ 

#### Box-Jenkins model identitification workflow (full ARIMA modelling process) 

1. Plot the data and identify unusual observations, look at patterns, trends, seasonality, outliers, is it stationary? 
2. Use a Box-Cox transformation to stabilize the variance if necessary. If not, move to step 6
3. After selecting the model order, difference the data if necessary, until it is stationary. Use unit-root tests to be sure. 
4. Plot the ACF and PACF of the differenced data to determine which model suits best
5. Try the chosen model and use the AICc to search for a better model
6. Check the residuals by plotting the ACF of them and doing a port-manteau test of the residuals
7. If the residuals do not look like white noise, repeat from step 4. If they do, calculate the forecasts

#### Forecasting 

- Rearange ARIMA equation so yt is on the LHS
- Replace t with T + h
- On RHS, replace future observations with their forecasts and future errors by 0 and past errors by corresponding residuals
- Start with h = 1 and repeat
