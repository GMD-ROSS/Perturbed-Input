### Project Steps

1. Read Hyndman & Athanasopoulos Chapters 2 and 3. Take notes on stationarity, differencing, and autocorrelation.
2. Read Chapter 9 (ARIMA models). Understand the Box-Jenkins model identification workflow.
3. Using pen and paper (or a Jupyter notebook), work through the ACF and PACF for a simple AR(1) process with known parameters.
4. Write a short summary (1–2 pages) explaining the difference between AR, MA, ARMA, and ARIMA models, and when each is appropriate.
5. Install the required Python libraries: `statsmodels`, `pmdarima`, `numpy`, `pandas`, `matplotlib`, `scipy`.

### Key Concepts to Master
- Stationarity and the Augmented Dickey-Fuller (ADF) test
- Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF)
- The Box-Jenkins identification → estimation → diagnostic checking cycle
- Information criteria for model selection: AIC and BIC
