## Chapter 2 - The Forecaster's Toolbox

### Covariance 
- Measures the extent of the linear relationship between two variables
- Negative covariance means they move in opposite directions, positive covariance they move together (direction) 

### Correlation 
- Standardized covariance between -1 and 1
- Measures both strength of the relationship and direction

### Autocovariance 
- Gives the covariance of a process with itself at pairs of time points
- Not standardized

$$
c_k = \frac{1}{T} \sum_{t=k+1}^{T} (y_t - \bar{y})(y_{t-k} - \bar{y})
$$

- Where k is the lag by rk

### Autocorrelation 
- How similar the signal is to itself after it is shifted by k units
- Standardized
- For uncorrelated data, the autocorrelation should be close to 0
- co is the variance 

$$
r_k = \frac{c_k}{c_0}
$$

### Correlogram

- A plot of ACF values vs lag

- Seasonal ACF is not cyclical although it typically has a calendar based pattern
- Cyclical ACF does not follow a calendary based patter, but repeats itself 

### Residuals 

- Difference between observed value and its forecast based on all previous observations (error) 
- Uncorrelated, meaning there should be no pattern in the errors (random)
- If the errors are correlated, the model may have missed something
- Residuals should have a mean of 0 otherwise they are biased 
- Constant variance, normal distribution

### White Noise 

- Uncorrelated across time with zero mean and constant variance
- Residuals should look like white noise, a good diagnostic is to check the ACF of residuals through Portmanteau tests, Box-Pierce tests, Ljung-Box tests
- If all autocorrelation values are within the critical value limits, the data set is white noise. Critical values are found by: 

$$
\pm \frac{1.96}{\sqrt{T}}
$$

### Forecast Accuracy 

Mean Absolute Error (MAE)
$$
\mathrm{MAE} = \frac{1}{T} \sum_{t=1}^{T} \left| y_t - \hat{y}_t \right|
$$

Mean Squared Error (MSE)
$$
\mathrm{MSE} = \frac{1}{T} \sum_{t=1}^{T} \left( y_t - \hat{y}_t \right)^2
$$

Root Mean Squared Error (RMSE)
$$
\mathrm{RMSE} = \sqrt{ \frac{1}{T} \sum_{t=1}^{T} \left( y_t - \hat{y}_t \right)^2 }
$$

Mean Absolute Percentage Error (MAPE)
$$
\mathrm{MAPE} = \frac{100}{T} \sum_{t=1}^{T} \left| \frac{y_t - \hat{y}_t}{y_t} \right|
$$

Mean Absolute Scaled Error (MASE)

$$
\mathrm{MASE} = \frac{1}{T} \sum_{t=1}^{T} \frac{|y_t - \hat{y}_t|}{Q}
$$

Q is a scaling factor.

$$
Q = \frac{1}{T-1} \sum_{t=2}^{T} |y_t - y_{t-1}|
$$

Seasonal Data
$$
Q = \frac{1}{T-m} \sum_{t=m+1}^{T} |y_t - y_{t-m}|
$$

- We want the error to be low for each of the error computations

## Chapter 3 - Exponential Smoothing

- Smoothing equation is:

$$
\ell_t = \alpha y_t + (1 - \alpha)\ell_{t-1}       
$$

- Forecase equation is the summation of all weighted past data: 
$$
\hat{y}_{t+h|t} = \sum_{j=1}^{t} \alpha (1-\alpha)^{t-j} y_j + (1-\alpha)^t \ell_0
$$
### State Space Form

#### Observation equation

- What we observe is what we expect + an error 

$$
y_t = \ell_{t-1} + e_t
$$

#### Computing the Error
$$
e_t = y_t - \hat{y}_t
$$

#### State Equation 

$$
\ell_t = \ell_{t-1} + \alpha e_t
$$

- Key idea is this: we predict $y_t$, observe a real value, subtract the prediction from the real value for an error and update the estimate
- Alpha and $l_o$ must be estimated, often by minimising the MSE

### Chapter 6 - Making Time Series Stationary 

- If the size of the time series fluctuations change, ie. variations at different levels, transformations will be needed
- This essentially makes the variance constant and suitable for ARIMA and ETS 
- Transformation usually does not make a difference to forecasts, but has an effect on PI (Prediction Intervals) ie. uncertainty range
   
#### Box-Cox Transformations 

- For $\lambda = 0,$
  $w_t = \log(y_t)$
- For $\lambda \ne 0,$
  $w_t = \frac{y_t^\lambda - 1}{\lambda}$

- Choose a simple value of lambda
- Reverse transformations must be done to obtain forecasts on the original scale
- For $\lambda = 0$,
  $y_t = \exp(w_t)$
- For $\lambda \ne 0$,
  $y_t = (\lambda w_t + 1)^{1/\lambda}$

- Box-Cox transformation accompanied with an additive (must be additive) ETS model is better than ETS model alone

#### Stationarity 

- Stationary series are horizontal, does not depend on t
- Constant variance, constant mean, no patterns predictable (trend or seasonality) think noise around a fixed level
- How do we identify non-stationary series?
- The ACF of stationary data drops to zero relatively quickly
- The ACF of non-stationary data decreases slowly
- Typically for non-stationary data, the value of r1 is often large and positive

#### Ordinary Differencing 

- Differencing is used to stabilize the mean (ARIMA model needs stationary data) 
- The differenced series is the changes between each observation in the original series

#### Second-Order Differencing 

- We can difference the data a second time if it is not stationary after a single difference

#### Seasonal Differencing 

- Seasonal differenced series are closer to being stationary
- Can be combined with first differencing

- First differences are the changes between one observation and the next
- Seasonal differences are the changes between one year to the next 

#### Unit Root tests 

- Does the data wonder over time? If it does, it has a unit root, and therefore we need to difference it, if not, its stationary
- Dickey-Fuller test (null that data is non-stationary and non-seasonal)
- Kwiatkowksi-Phillips-Schmidt-Shin test (null that data is stationary and non-seasonal)

### Chapter 7 - Non-Seasonal ARIMA models 

#### Autoregressive models (AR)   

$y_t = c + \phi_1 y_{t-1} + \phi_2 y_{t-2} + \cdots + \phi_p y_{t-p} + e_t$

- $e_t$ is white noise

##### AR(1) model 

$y_t = c + \phi_1 y_{t-1} + e_t$

- Current value depends on last value and some random noise
- $\phi_1$ should be between -1 and 1
- If p > 1 (number of lagged values in model) , $\phi_2$ should be between -1 and 1, $\phi_1 + \phi_2 < 1$ , $\phi_2 - \phi_1 < 1$
  
#### Moving average models (MA) 

$y_t = c + e_t + \theta_1 e_{t-1} + \theta_2 e_{t-2} + \cdots + \theta_q e_{t-q}$

- The prediction is made from a linear combination of past errors 

#### ARIMA models 

##### Autoregressive Moving Average models (ARMA) 

$y_t = c + \phi_1 y_{t-1} + \cdots + \phi_p y_{t-p} + \theta_1 e_{t-1} + \cdots + \theta_q e_{t-q} + e_t$ 

- Predictors include lagged values of $y_t$ and lagged errors
- Conditions on coefficients ensure invertibility and stationarity 

##### Autoregressive Integrated Moving Average models (ARIMA)

- These combine ARMA model with differencing
  
- Backshift operators:

$
B y_t = y_{t-1}
$

$
B^2 y_t = y_{t-2}
$

- Substituting:
  
$
y_t = c + \phi_1 B y_t + \cdots + \phi_p B^p y_t
      + \theta_1 B e_t + \cdots + \theta_q B^q e_t + e_t
$

- Factor yt:

$
y_t - (\phi_1 B + \cdots + \phi_p B^p)y_t = c + (1 + \theta_1 B + \cdots + \theta_q B^q)e_t
$

- Simplifying: 

$
(1 - \phi_1 B - \cdots - \phi_p B^p)y_t = c + (1 + \theta_1 B + \cdots + \theta_q B^q)e_t
$

- Polynomial for phi

$
\phi(B) = 1 - \phi_1 B - \cdots - \phi_p B^p
$

- Polynomial for theta 

$
\theta(B) = 1 + \theta_1 B + \cdots + \theta_q B^q
$

- ARMA form: 

$
\phi(B) y_t = c + \theta(B) e_t
$

- By applying differencing, we transform to the ARIMA model:

$
\phi(B)(1 - B)^d y_t = c + \theta(B) e_t
$  

##### ARIMA(p,d,q) model 

- AR: p = order of autoregressive part
-  I: d = degree of first differencing involved
- MA: q = order of the moving average part
- White noise model: ARIMA(0,0,0)
- Random walk: ARIMA(0,1,0) with no constant
- Random walk with drift: ARIMA(0,1,0) with constant

- The higher the value of d, the more rapidly Forecast intervals increase in size
- For cyclic forecasts, p > 2 and restrictions on the coefficients are asserted
- For p = 2: 
$\phi_1^2 + 4\phi_2 < 0$ 

##### Partial autocorrelations 

- Parial autocorrelations measure the relationship between one data point and another when the affects of other time lags are removed
- ACF: How related are the datapoints overall
- PACF: Is there a direct connection between a set of datapoints

#### Estimation and order Selection

##### Maximum likelihood estimation (MLE) 

- Once we identify the model order, we need to estimate parameters c, $\phi_p$, $\theta_p$

#### Information Criteria

##### Akaike's Information Criterion (AIC): 
$
\mathrm{AIC} = -2 \log(L) + 2(p + q + k + 1)
$

- p is the number of AR terms (how many lagged y's)
- q is the number of MA terms (how many lagged errors)
- L is the likelihood of the data
- k=1 if c$\neq$0 and k = 0 if c = 0
- T is the number of observations

##### Corrected AIC: 

$
\mathrm{AIC}_c = \mathrm{AIC} + \frac{2(p + q + k + 1)(p + q + k + 2)}{T - p - q - k - 2}
$

##### Bayesian Information Criterion: 

$
\mathrm{BIC} = \mathrm{AIC} + (p + q + k + 1)\log(T)
$ 

#### Box-Jenkins model identitification workflow (full ARIMA modelling process) 

1. Plot the data and identify unusual observations, look at patterns, trends, seasonality, outliers, is it stationary? 
2. Use a Box-Cox transformation to stabilize the variance if necessary. If not, move to step 6
3. After selecting the model order, difference the data if necessary, until it is stationary. Use unit-root tests to be sure. 
4. Plot the ACF and PACF of the differenced data to determine which model suits best
5. Try the chosen model and use the AICc to search for a better model
6. Check the residuals by plotting the ACF of them and doing a port-manteau test of the residuals
7. If the residuals do not look like white noise, repeat from step 4. If they do, calculate the forecasts

#### Forecasting 

- Rearange ARIMA equation so yt is on the LHS
- Replace t with T + h
- On RHS, replace future observations with their forecasts and future errors by 0 and past errors by corresponding residuals
- Start with h = 1 and repeat


### Differences between AR,MA, ARMA and ARIMA models and their best uses 

#### AR (Autoregressive) 

The idea behind AR is that the current estimation depends on past values. It's best used when the data is already stationary, meaning it has no trend. The ACF for AR decays gradually because it has a strong correlation with past values. Used when we have a smooth dependancy on past values, and a gradual decay in the ACF. 

#### MA (Moving Average) 

The idea behind Ma is that the current estimation depends on past errors. It essentially is working to correct past mistakes. Ideal for stationary data, and the ACF function cuts off quickly. An example of this is when a stock reacts to recent unexpected news. Used when the data is unpredictable, and the ACF dies fast. 

#### ARMA (Combination of Moving Average and Autoregressive) 

ARMA uses both past values and errors to calculate the prediction. Used when data is stationary and complex. 

#### ARIMA 

Since most real data is non-stationary and often trending, we need to difference the data to make it stationary in order to use ARMA. This is where ARIMA comes in handy, it transforms the data into an appropriate format for ARMA. We remove the trend to make it stationary through differencing, then apply ARMA. Examples would be stock prices, GDP. 




### VARMA 

Let $y_t$ be a stationary d-dimensional mean zero vector time series. It follows a $VARMA_d$(p,q) model if: 

$y_t = \sum_{l=1}^{p} \Phi_l y_{t-l} + \sum_{m=1}^{q} \Theta_m a_{t-m} + a_t$

where $\Phi_l$ are the autoregressive (AR) parameter matrices, and $\Theta_m$ are teh moving average (MA) parameter matrices. $a_t$ dennotes a d-dimensional mean zero white noise vector time series. The VARMA model states that $y_t$ is a function of its own p past values and q lagged error terms. This model can therefore be written as: 

$\Phi(L) y_t = \Theta(L) a_t$


$\Phi(L) = I - \Phi_1 L - \Phi_2 L^2 - \cdots - \Phi_p L^p$

$\Theta(L) = I + \Theta_1 L + \Theta_2 L^2 + \cdots + \Theta_q L^q$

The lag operator $L^l$ is defined as $L^ly_t$ = $y_{t-l}$ , I denotes the dxd identity matrix. 

The VARMA model is stable if the determinant of $\Phi_z$ is not 0 for all |z| <1 and its invertible if the deterimnant of $\Theta_z$ is not equal to 0 for all |z| <1 . It is unit root stationary if the determinant of $\Phi_z$ is 0 for z = 1. 

https://wires.onlinelibrary.wiley.com/doi/10.1002/wics.70009

### VARIMA 

If $y_t$ is a stationary d-dimensional mean zero vector time series, it follows a $VARMDA_d$ (p,q) model if : 

$y_t = \sum_{l=1}^{p} \Phi_l y_{t-l} + \sum_{m=1}^{q} \Theta_m a_{t-m} + a_t$

where $\Phi_l$ is the autoregressive (AR) parameter matrices, and $\Theta_m$ is the moving average (MA) parameter matrices. $a_t$ denotes a mean zero whitenoise vector time series with d x d matrix. The VARMA model states that $y_t$ is
a function of its own mean past values p and q lagged error terms. 

It can be written as: 

$\Phi(L) y_t = \Theta(L) a_t$

where $a_t$ is noise

The model is stable if the determinant of $\Phi_z$ is not equal to 0, for all |z| < 1 , and is invertible if the determinant of $\Theta_z$ is not equal to zero for all |z| < 1 . It is unit root stationary if the 
determinant of $\Phi_z$ = 0 for some |z| = 1 . If it is invertible, it can be written with a VAR representation as :

The issue with VARIMA models is that the AR and MA are not cleanly seperable, so multiple variations of the same output can exist. 

With ARIMA, we ensure uniqueness in a arima model by ensuring the MA and AR polynomials share no common factors, secondly, the MA polynomial must be invertible (roots lie outside the unit circle). We cant easily do this with matrices.

### VAR 

- The VAR model is a generalization of the univariate autoregressive model for forecasting a vector of time series. It has one equation per variable in the system. For a 
2 dimensional VAR(1), we have: 

$
y_{1,t} = c_1 + \phi_{11,1} y_{1,t-1} + \phi_{12,1} y_{2,t-1} + e_{1,t}
$

$
y_{2,t} = c_2 + \phi_{21,1} y_{1,t-1} + \phi_{22,1} y_{2,t-1} + e_{2,t}
$

where e1t and e2t are white noise processes. The coefficient of $\Phi_ii$ represents the influence of the lth lag of variable yi on itself, whereas the coefficient $\Phi_ij$ captures the 
influence of the lth variable of yj on yi. 

- If the series is stationary, we difference the data in order to make it stationary, then fit a VAR model. The models are estimated by equation using the principle of least squares. 
- Forecasts are generated from a VAR model in a recursive manner, where the VAR generates forecasts for each variable included in the system. 

- There are two decisions to make when using a VAR to forecast, namely how many variables (denoted by K) and how many lags (denoted by p). 
The number of coefficients to be estimated in a VAR is equal to K + pK^2 or, 1 + pK per equation
- The more coefficients that need to be estimated, the higher the estimation error
- In practice, we keep the K value small and only include variables that are correlated to one another 
- Information criteria are commonly used to select the number of lags to be included ( AIC, HQ, SC and FPE)
- Be careful when using AIC, it tends to use the largest number of lags, typically BIC is chosen

- VARS are generally useful for:
1. Forecasting a collection of related variables where no explicit interpretation is required
2. Testing whether one variable is useful in forecasting another (Granger causality tests)
3. Impulse response analysis, where the response of one variable to a sudden but temporary change in another is analyzed
4. Forecast error variance decomposition, where the proportion of the forecast variance of each variable is attributed to the effects of the other variables 

#### The stationary vector autoregression model 

Let Yt denote a nx1 vector of time series variables. The basic p lag vector autoregressive VAR (p) model has the form:

$Y_t = c + \Pi_1 Y_{t-1} + \Pi_2 Y_{t-2} + \cdots + \Pi_p Y_{t-p} + \varepsilon_t$

where $\Pi_i$ are n x n coefficiant matrices and $\varepsilon_t$ is a n x 1 zero mean unobservable mean white noise vector process with covariance matrix $\sum$. 

Example, the bivariate VAR(2) model equation has the form: 

$$
\begin{bmatrix}
y_{1,t} \\
y_{2,t}
\end{bmatrix}
=
\begin{bmatrix}
c_1 \\
c_2
\end{bmatrix}
+
\begin{bmatrix}
\phi_{11}^{(1)} & \phi_{12}^{(1)} \\
\phi_{21}^{(1)} & \phi_{22}^{(1)}
\end{bmatrix}
\begin{bmatrix}
y_{1,t-1} \\
y_{2,t-1}
\end{bmatrix}
+
\begin{bmatrix}
\phi_{11}^{(2)} & \phi_{12}^{(2)} \\
\phi_{21}^{(2)} & \phi_{22}^{(2)}
\end{bmatrix}
\begin{bmatrix}
y_{1,t-2} \\
y_{2,t-2}
\end{bmatrix}
+
\begin{bmatrix}
\varepsilon_{1,t} \\
\varepsilon_{2,t}
\end{bmatrix}
$$

Expanded: 

$$
y_{1,t} = c_1
+ \phi_{11}^{(1)} y_{1,t-1}
+ \phi_{12}^{(1)} y_{2,t-1}
+ \phi_{11}^{(2)} y_{1,t-2}
+ \phi_{12}^{(2)} y_{2,t-2}
+ \varepsilon_{1,t}
$$

$$
y_{2,t} = c_2
+ \phi_{21}^{(1)} y_{1,t-1}
+ \phi_{22}^{(1)} y_{2,t-1}
+ \phi_{21}^{(2)} y_{1,t-2}
+ \phi_{22}^{(2)} y_{2,t-2}
+ \varepsilon_{2,t}
$$

where $\operatorname{cov}(\varepsilon_{1,t}, \varepsilon_{2,s}) = \sigma_{12}$

In lag operator notation, the VAR(p) is written as: 

$
\Pi(L)Y_t = c + \varepsilon_t
$

where: 

$
\Pi(L) = I_n - \Pi_1L - \cdots - \Pi_pL^p
$

The VAR(p) is stable if the roots of $ \det\left(I_n - \Pi_1z - \cdots - \Pi_p z^p\right)=0$ = 0 lie outside the complex unit circle, or if the eigen values of the companion matrix

$
F =
\begin{bmatrix}
\Pi_1 & \Pi_2 & \cdots & \Pi_p \\
I_n & 0 & \cdots & 0 \\
0 & I_n & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & I_n & 0
\end{bmatrix}
$

have a modulus less than 1. 

If Yt is covariance stationary, the mean is given by: 

$
\mu = \left(I_n - \Pi_1 - \cdots - \Pi_p\right)^{-1} c
$

and therefore the mean adjusted form of VAR(p) is: 

$Y_t - \mu = \Pi_1 (Y_{t-1} - \mu) + \Pi_2 (Y_{t-2} - \mu) + \cdots + \Pi_p (Y_{t-p} - \mu) + \varepsilon_t$

Since the basic VAR(p) model may be too restrictive to represent the main characteristics of the data, we may need to add deterministic terms such as a linear time trend or seasonal
dummy variables. The general form of VAR(p) with these variables integrated is: 

$ Y_t = \Pi_1 Y_{t-1} + \Pi_2 Y_{t-2} + \cdots + \Pi_p Y_{t-p} + \Phi D_t + G X_t + \varepsilon_t$

where Dt represents a l x 1 matrix of deterministic components, Xt represents an m x 1 matrix of exogenous variables, and phi and G are parameter matrices.
